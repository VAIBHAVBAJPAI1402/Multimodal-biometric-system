```python
import cv2
import time
```

```python
img = cv2.imread('1.jpeg',1)
```

```python
# resized = cv2.resize(img, (900,600))
resized =cv2.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)))

#building cascade classifier object
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

#read image as a grey scale image
gray_img = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray_img,scaleFactor=1.3, minNeighbors=5)

print(type(faces))

for x,y,w,h in faces:
    img_out = cv2.rectangle(resized, (x,y), (x+w,y+h), (0,255,0), 3)

cv2.imshow("Face detected", img_out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture(0)
#check shows whether python is able to capture the frame
#frame is numpy array and stores first image that video captures
check, frame = video.read()
print(check)
print(frame)

time.sleep(3)
cv2.imshow("Capturing",frame)
cv2.waitKey(0)
video.release()
cv2.destroyAllWindows()
```

```python

#gray scale video capturing
import cv2,time
cam = cv2.VideoCapture(-1)
a = 1

while True:
    a=a+1
    check, frame = cam.read()
    print(check)
    print(frame)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("Capture",gray)
    key = cv2.waitKey(1)
    if key== ord('q'):
        break    
print(a)
cam.release()
cv2.destroyAllWindows()
```
